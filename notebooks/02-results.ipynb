```python
import pandas as pd
```

```python
results = pd.read_csv("../data/training_metrics_20251203_150830.csv")
```

```python
results
```

```python
import matplotlib.pyplot as plt

fig, axes = plt.subplots(2, 2, figsize=(14, 10))
fig.suptitle("Training and Test Metrics Over Epochs", fontsize=16)

# Epoch vs Train Loss
axes[0, 0].plot(results["epoch"], results["train_loss"], label="Train Loss", color="tab:blue")
axes[0, 0].set_title("Train Loss")
axes[0, 0].set_xlabel("Epoch")
axes[0, 0].set_ylabel("Loss")
axes[0, 0].grid(True)

# Epoch vs Train Accuracy
axes[0, 1].plot(results["epoch"], results["train_accuracy"], label="Train Accuracy", color="tab:green")
axes[0, 1].set_title("Train Accuracy")
axes[0, 1].set_xlabel("Epoch")
axes[0, 1].set_ylabel("Accuracy")
axes[0, 1].grid(True)

# Epoch vs Test Loss
axes[1, 0].plot(results["epoch"], results["test_loss"], label="Test Loss", color="tab:red")
axes[1, 0].set_title("Test Loss")
axes[1, 0].set_xlabel("Epoch")
axes[1, 0].set_ylabel("Loss")
axes[1, 0].grid(True)

# Epoch vs Test Accuracy
axes[1, 1].plot(results["epoch"], results["test_accuracy"], label="Test Accuracy", color="tab:purple")
axes[1, 1].set_title("Test Accuracy")
axes[1, 1].set_xlabel("Epoch")
axes[1, 1].set_ylabel("Accuracy")
axes[1, 1].grid(True)

plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```

The model clearly overfits given the divergence of the loss curves, but for the exercise it's enough
